```python
import numpy as np
import pandas as pd
from IPython.display import display
```

```python
func = lambda x: x**2 + np.log(x)
abs_error = lambda x, y: abs(x - y) if x else "inf"
nodes = [round(0.4 + i * (0.9 - 0.4)/10, 2) for i in range(11)]
values = [func(i) for i in nodes]
print(nodes, values)
```

```python
def compute_central_differences(values: list[float]) -> list[list[float]]:
    """
    Функция, которая считает центральные конечные разности

    Args:
        values (list[float]): Список известных значений функции

    Returns:
        list[list[float]]: список конечных разностей, где первый индекс означает порядок разности, а второй то, от каких значений мы ее получаем
        Например differences[1][2] - означает конечную разность первого порядка от элементов 2 и 3 в 0-индексации
    """
    differences = [values]
    while len(differences[-1]) > 1:
        current_diff = [
            differences[-1][i + 1] - differences[-1][i]
            for i in range(len(differences[-1]) - 1)
        ]
        
        differences.append(current_diff)
    return differences
```

```python
all_differences = compute_central_differences(values)
print(*all_differences, sep='\n')
```

Первая формула Ньютона по $x_0, x_1, x_2, ..., x_n = x_0 + th, \ 0 < t < 1$

$\displaystyle P_n(x) = a_0 + a_1(x - x_0) + a_2(x - x_0)(x - x_1) + ... + a_n(x - x_0)(x - x_1)...(x - x_{n-1}), \ a_k = \frac{\Delta^ky_0}{k!h^k}$

```python
def first_newton_interpolate(n: int, node: float, x_0: float, h: float) -> float:
    """ Первая интерполяционная функция Ньютона 

    Args:
        n (int): количество узлов
        node (float): точка, в которой нужно вычислить значение
        x_0 (float): первый узел
        h (float): шаг между узлами

    Returns:
        float: предположительное значение функции в точке node
    """
    ans = func(x_0)
    diffs = 1
    factorial = 1
    for i in range(1, n + 1):
        factorial *= i
        diffs *= node - (x_0 + (i - 1)*h)
        ans += all_differences[i][0] / (factorial *  h**i) * diffs
    return ans
```

```python
need_check = [0.52, 0.42, 0.87, 0.67]
n = 10
x_0 = 0.4
h = 0.05
dataframe = pd.DataFrame(columns=["Точка интерполирования", "многочлен Ньютона", "истинное значение", "абсолютная ошибка"])

for i in range(4):
    ni = first_newton_interpolate(n, need_check[i], x_0, h)
    f = func(need_check[i])
    dataframe.loc[len(dataframe)] = [need_check[i], ni, f, abs_error(ni, f)]

display(dataframe)
```

Вторая интерполяционная формула Ньютона

$\displaystyle P_n(x) = a_0 + a_1(x - x_n) + a_2(x - x_n)(x - x_{n-1}) + ... + a_n(x - x_n)(x - x_{n-1})...(x - x_1), \ a_k = \frac{\Delta^ky_{n-k}}{k!h^k}$

```python
def second_newton_interpolate(n: int, node: float, x_0: float, h: float) -> float:
    """Вторая интерполяционная формула Ньютона

    Args:
        n (int): количество узлов
        node (float): точка, в которой нужно вычислить значение
        x_0 (float): первый узел
        h (float): шаг между узлами

    Returns:
        float: предположительное значение функции в точке node
    """
    ans = func(x_0 + n * h)
    diffs = 1
    factorial = 1
    for i in range(1, n + 1):
        factorial *= i
        diffs *= node - (x_0 + (n - i + 1)*h)
        ans += all_differences[i][-1] / (factorial *  h**i) * diffs
    return ans
```

```python
need_check = [0.52, 0.42, 0.87, 0.67]
n = 10
x_0 = 0.4
h = 0.05
dataframe = pd.DataFrame(columns=["Точка интерполирования", "многочлен Ньютона", "истинное значение", "абсолютная ошибка"])

for i in range(4):
    ni = second_newton_interpolate(n, need_check[i], x_0, h)
    f = func(need_check[i])
    dataframe.loc[len(dataframe)] = [need_check[i], ni, f, abs_error(ni, f)]

display(dataframe)
```

```python
def find_nearest(grid: list[float],
                 node: float = 0) -> int:
    """
    Эта функция возвращает индекс ближайшего по значению элемента из списка grid к числу node
    Args:
        grid (list[float]): Список узлов сетки. Defaults to None.
        node (float): Точка, к которой нужно найти ближайюшую из сетки. Defaults to 0.

    Returns:
        int: Индекс ближайшего элемента
    """
    l = 0
    r = len(grid) - 1
    while r - l > 1:
        avr = (r + l) // 2
        if grid[avr] > node:
            r = avr
        else:
            l = avr
    return l 
```

Первая формула Гаусса

$\displaystyle P_{2m}(x) = a_0 + a_1(x - x_k) + a_2(x - x_k)(x - x_{k + 1}) + a_3(x - x_k)(x - x_{k+1})(x - x_{k-1}) + ... + a_{2m}(x - x_k)(x - x_{k+1})...(x - x{k + m}), \\ a_{2n} = \frac{\Delta^{2n}y_{-n}}{2n! \cdot h^{2n}}$

Здесь $x_k$ - ближайший узел к $x$

Имеет место замена $q = \frac{x - x_0}{h}$

```python
def gauss_interpolate(n: float, node: float, h: float, type = 'forward') -> float:
    """Интерполяционная формула Гаусса

    Args:
        n (float): количество узлов
        node (float): точка, в которой нужно вычислить значение
        h (float): шаг между узлами
        type (str, optional): Тип интерполяции (Как и у Ньютона тут две формулы, я объединил их в одну функцию). Defaults to 'forward'.

    Returns:
        float: предположительное значение функции в точке node 
    """
    start_node = find_nearest(nodes, node)
    q = (node - nodes[start_node]) / h
    ans = values[start_node]
    diffs = 1
    factorial = 1

    for k in range(1, n + 1):
        d = (k + 1) // 2
        i = (start_node - d) if type == 'forward' else (start_node - k // 2)
        diffs *= (q - d) if not k % 2 else (q + d)
        factorial *= k
        if len(all_differences[k]) > abs(i):
            ans += (diffs / factorial) * all_differences[k][i]
        else: 
            break
    return ans
```

```python
need_check = [0.52, 0.42, 0.87, 0.67]
n = 10
x_0 = 0.4
h = 0.05
type = 'forward'
dataframe = pd.DataFrame(columns=["Точка интерполирования", "многочлен Гаусса", "истинное значение", "абсолютная ошибка"])

for i in range(4):
    ni = gauss_interpolate(n, need_check[i], x_0, h, type)
    f = func(need_check[i])
    dataframe.loc[len(dataframe)] = [need_check[i], ni, f, abs_error(ni, f)]

display(dataframe)
```

Формула Стирлинга представляет собой среднее арифметическое между двумя формулами Лагранжа.

$\displaystyle P(x) = y_0 + q\frac{\Delta y_{-1} + \Delta y_0}{2} + \frac{q^2}{2}\Delta^2y_{-1} + \frac{q(q^2 - 1^2)}{3!} \cdot \frac{\Delta^3 y_{-2} + \Delta^3 y_{-1}}{2} + ... + \frac{q(q^2 - 1^2)(q^2 - 2^2) ... (q^2 - (n - 1)^2)}{2n!}\Delta^{2n}y_{-n}$

$\displaystyle q = \frac{x - x_0}{h}$

```python
def stirling_interpolate(n: float, node: float, h: float) -> float:
    """Интерполяционная формула Стирлинга

    Args:
        n (float): количество узлов
        node (float): точка, в которой нужно вычислить значение
        h (float): шаг между узлами

    Returns:
        float: предположительное значение функции в точке node
    """
    start_node = find_nearest(nodes, node)
    q = (node - nodes[start_node]) / h
    ans = values[start_node]
    diffs = q
    factorial = 1

    for k in range(1, n + 1):
        factorial *= k
        i = start_node - (k // 2)
        if len(all_differences[k]) > abs(i):
            if k % 2:
                delta_k = (all_differences[k][i-1] + all_differences[k][i]) / 2
                diffs *= q**2 - k**2 if k > 1 else 1
            else:
                delta_k = all_differences[k][i]
                diffs *= q
        else:
            break
        ans += (diffs / factorial) * delta_k
    return ans
```

```python
need_check = [0.52, 0.42, 0.87, 0.67]
n = 10
x_0 = 0.4
h = 0.05
dataframe = pd.DataFrame(columns=["Точка интерполирования", "многочлен Стирлинга", "истинное значение", "абсолютная ошибка"])

for i in range(4):
    ni = stirling_interpolate(n, need_check[i], h)
    f = func(need_check[i])
    dataframe.loc[len(dataframe)] = [need_check[i], ni, f, abs_error(ni, f)]

display(dataframe)
```

Интерполяционная формула Бесселя

\begin{aligned}
            P_n(x_m + th) &= \frac{f(x_m) + f(x_{m+1})}{2} + \\
            &+ (t - \frac{1}{2}) \Delta^1 f(x_m) + \\
            &+ \frac{t(t-1)}{2!} \left[\frac{\Delta^2 f(x_{m-1}) +
            \Delta^2 f(x_m)}{2}\right] + \\
            &+ \ldots + \\
            &+ \frac{t(t^2-1)\ldots\left[t^2-(n-1)^2\right](t-n)}{(2n)!}
            \left[\frac{\Delta^{2n} f(x_0) + \Delta^{2n} f(x_1)}{2}\right] + \\
            &+ \frac{t(t^2-1)\ldots\left[t^2-(n-1)^2\right](t-n)
            (t-\frac{1}{2})}{(2n+1)!} \Delta^{2n+1} f(x_0)
        \end{aligned}


```python
def bessel_interpolate(n: float, node: float, h: float) -> float:
    """Интерполяционная формула Бесселя

    Args:
        n (float): количество узлов
        node (float): точка, в которой нужно найти значение
        h (float): шаг между узлами

    Raises:
        ValueError: вызывается при четном числе узлов, так как иначе будет сложно двигаться симметрично

    Returns:
        float: предположительное значение в точке node
    """

    if n % 2 == 0:
        raise ValueError("n must be odd")
    start_node = n//2
    t = (node - nodes[start_node]) / h   

    ans = (all_differences[0][start_node] + all_differences[0][start_node + 1]) / 2
    factorial = 1
    diffs = 1

    for k in range(1, n + 1):
        i = start_node - (k // 2)
        factorial *= k
        if k % 2 == 1:
            delta_k = all_differences[k][i]
            p = diffs * (t - 1 / 2) * delta_k
        else:
            delta_k = (all_differences[k][i] + all_differences[k][i + 1]) / 2
            diffs = 1
            for j in range(1, k // 2):
                diffs *= t**2 - j**2
            diffs *= t * (t - k // 2)
            p = diffs * delta_k
        term = p / factorial
        ans += term

    return ans
```

```python
need_check = [0.52, 0.42, 0.87, 0.67]
n = 10
x_0 = 0.4
h = 0.05
dataframe = pd.DataFrame(columns=["Точка интерполирования", "многочлен Бесселя", "истинное значение", "абсолютная ошибка"])

for i in range(4):
    ni = bessel_interpolate(9, need_check[i], h)
    f = func(need_check[i])
    dataframe.loc[len(dataframe)] = [need_check[i], ni, f, abs_error(ni, f)]

display(dataframe)
```
